````python
import os
from groq import Groq

GROQ_API_KEY = os.environ['GROQ_API_KEY']

client = Groq(api_key=GROQ_API_KEY)

model = 'llama-3.1-8b-instant'
````

````python
def get_response(prompt):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content
````

````python
# Effective verbs: Write, Complete, Explain, Describe, Evaluate, ...

# Non-effective verbs: Understand, Think, Feel, Try, Know, ... (verbos ambiguos)
````

````python
# Few-shot prompting
system_prompt = """
Your task is to perform a sentiment analysis on customer reviews.
You are assigning a number for a given customer conversation:
-1 if the sentiment is negative, 1 if positive.
"""

response = client.chat.completions.create(
  model = "llama-3.3-70b-versatile",
  # Provide the examples as previous conversations
  messages = [{"role": "system", "content": system_prompt},
              {"role": "user", "content": "The product quality exceeded my expectations"},
              {"role": "assistant", "content": "1"},
              {"role": "user", "content": "I had a terrible experience with this product's customer service"},
              {"role": "assistant", "content": "-1"},
              # Provide the text for the model to classify
              {"role": "user", "content": "The price of the product is really fair given its features."}
             ],
  temperature = 0
)
print(response.choices[0].message.content)
````

````python
# System prompts

def get_response(system_prompt, user_prompt):
  # Assign the role and content for each message
  messages = [{"role": "system", "content": system_prompt},
      		  {"role": "user", "content": user_prompt}]  
  response = client.chat.completions.create(
      model=model, messages= messages, temperature=0)
  
  return response.choices[0].message.content

# Try the function with a system and user prompts of your choice 
response = get_response("You are an expert Data Science Tutor that responds concretely.", "How can I automate an ARIMA model in python?")
print(response)
````
